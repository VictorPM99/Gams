```
num=0
cont=0
list=[]
while num in range(0,5000000):
  num=num+1
  if num%2==0 and num%37==0 and num%49==0:
    cont=cont+1
    list.append([num])

print(f'O total de números no intervalo de 1 a 5 milhões, pares, múltiplos de 49 e múltiplos de 37 são: {cont}')
print(f'Os números são: {list}')
```
